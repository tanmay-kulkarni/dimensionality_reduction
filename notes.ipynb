### Text Processing

If two documents are similar to each other, their corresponding vectors in the n-dimensional space are geometrically closer to each other. Some popular techniques for representing text as vectors are:
* Bag of Words (BoW)
* TFIDF (Term frequency inverse document frequency)
* W2V (Word to Vector)
* Avg. W2V (Average word to vector)
* TFIDF-W2V

**Bag of Words**

1. Construct a dictionary. The dictionary is basically the set of all unique words in your document.
2. Each word is a different dimension
3. Objective is that similar text should result in closer vectors

IDF - Inverse Document frequency

$N$ = Total number of documents

### KNN


Time and space Complexity of KNN = $O(n*d)$

Where,

$n$=number of data points

$d$=number of dimensions

* As k increases, the decision boundary becomes smoother
* For k = $n$, it will be a classifier that classifies everything in the majority class
* To see if KNN is underfitting or overfitting, plot k(x axis) vs training/validation error or accuracy(y axis)
    - if train error and validation errors are both high => Underfitting
    - if train error is low and validation error are high => Overfitting
    
*Locality Sensitive Hashing(LSH)* is a state of the art technique for reducing the time complexity of KNN 

### Weighted KNN

* The idea is to give different weights to neighbors based on how far or near they are, instead of considering all neighbors as equals.
* As the distance decreases, weight increases. As distance increases, weight decreases.
* To classify a point, calculate the weighted sum of all neighbors and assign the class that has more weight.
* Weight can be euclidian or some other



**Time based splitting of Data**

- sort the data in ascending order of time. If timestamp is not available, time based splitting is not possible obviously.
- Oldest 60% becomes train data. Next 20% validation and final 20% is Test data.
- For Amazon food reviews, time based splitting is preferable over random splitting. Because with time, the reviews change.
    and for random splitting, you probably won't get the same accuracy for new data. The accuracy you would get in time based splitting would be more indicative of the accuracy you are expected to get when you finally deploy your model in production.

### KNN using KD-Tree

Time complexity of KNN using KD-Tree

Best case:

$O(k*log(n))$

Worst case:

$O(k*n)$



**Limitations of KD-Tree**

* okay for small dimensions, but not suitable for high dimensional data

### Locality sensitive Hashing (LSH)

* Works very well even for high dimensional data.
* LSH is a randomized algorithm
